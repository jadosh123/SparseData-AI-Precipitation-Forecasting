```python
import numpy as np

def get_wind_components(ws, wd):
    """
    Converts Wind Speed (m/s) and Direction (deg) into U and V vectors.
    """
    wd_rad = np.deg2rad(wd)
    u = -ws * np.sin(wd_rad)
    v = -ws * np.cos(wd_rad)
    return u, v

```
